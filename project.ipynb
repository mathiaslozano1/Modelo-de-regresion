```python
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt
```

```python
datos = pd.read_csv('archive\supermarket_sales -.csv')
datos.head()
```

```python
Market = pd.DataFrame(datos)
Market
```

```python
city = Market['City']
city
```

```python
grupo = Market.groupby('Product line').sum()
grupo
plt.Figure(figsize=(10,10))
plt.barh(grupo.index, grupo['Quantity'])
plt.xlabel('Product line')
plt.ylabel('Quantity')
plt.title('Product line vs Quantity')
plt.show()
```

```python
total_quantity =  Market.groupby('City')['Quantity'].sum()

total_quantity
```

```python
plt.figure(figsize=(10,5))
plt.barh(total_quantity.index, total_quantity.values)
plt.title('Total quantity of products sold by city')
plt.xlabel('City')
plt.ylabel('Total quantity')
plt.show()
```

```python
genero=Market.groupby('Gender')['Quantity'].sum()
plt.figure(figsize=(10,5))  
plt.bar(genero.index, genero.values)
plt.title('Total quantity of products')
plt.show()
```

```python
valor_t = int(Market['Total'].sum())
valor_t
```

```python
valor=Market.groupby('Invoice ID')['Total'].sum()
valor
plt.figure(figsize=(10,5))
plt.bar(valor.index, valor.values)
plt.title('Total sales per invoice')
plt.show()
```

```python
valor_600 = Market[Market['Total'] > 1000]
valor_600
plt.figure(figsize=(10,5))  
plt.barh(valor_600['Invoice ID'], valor_600['Total'])
plt.title('Sales greater than 900')
plt.show()
```
